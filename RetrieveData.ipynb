```python
import opendatasets as od
import pandas as pd
import requests 
import os
import zipfile

def extract_dataframe_from_url_zip(csv_file, url):
    r = requests.get(url, stream=True)
    with open(os.path.join(os.getcwd(), "tmp.zip"), 'wb') as fd:
        for chunk in r.iter_content(chunk_size=128):
            fd.write(chunk)

    with zipfile.ZipFile(os.path.join(os.getcwd(), "tmp.zip"), 'r') as zip_ref:
        zip_ref.extractall(os.getcwd())

    dataframe = pd.read_csv(csv_file)

    os.remove(csv_file)
    os.remove('tmp.zip')

    return dataframe

def extract_dataframe_from_kaggle(csv_url, csv_file_name, folder_name):

    path = os.path.join(os.path.join(os.getcwd(), folder_name), csv_file_name)

    print(path)

    od.download(csv_url)

    dataframe = pd.read_csv(path)
    
    try:
        os.remove(os.path.join(os.getcwd(), folder_name))
    except PermissionError:
        print("CSV folder was not removed due to permissions. Path:", path)
    
    return dataframe
```

```python
extract_dataframe_from_kaggle(
    "https://www.kaggle.com/fedesoriano/company-bankruptcy-prediction/version/2?select=data.csv",
    "data.csv",
    "company-bankruptcy-prediction")
```
